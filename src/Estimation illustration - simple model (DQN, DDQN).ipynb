```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('..//data//ddqn_estimation_beta_L_simple_model.pkl', 'rb') as f:
    ddqn_estimation = pickle.load(f)
    
with open('..//data//dqn_estimation_beta_L_simple_model.pkl', 'rb') as f:
    dqn_estimation = pickle.load(f)
```

```python
size_multiplier = 0.7
FIGSIZE = (14*size_multiplier, 8*size_multiplier)

f, ax = plt.subplots(1,1, figsize=FIGSIZE)


dqn_index = np.argmin(dqn_estimation['log_mse'])
ddqn_index = np.argmin(ddqn_estimation['log_mse'])

ax.scatter([dqn_estimation['beta_L'][dqn_index]], [dqn_estimation['log_mse'][dqn_index]], 
           s=75, color='black', alpha=0.3)
ax.scatter([ddqn_estimation['beta_L'][ddqn_index]], [ddqn_estimation['log_mse'][ddqn_index]], 
           s=75, color='black', alpha=0.3, label='Optimal Beta_L')


ax.plot(dqn_estimation['beta_L'], dqn_estimation['log_mse'], label='DQN')
ax.plot(ddqn_estimation['beta_L'], ddqn_estimation['log_mse'], label='Double DQN', ls='--')


ax.set_xlabel('beta L')
ax.set_ylabel('log(mse)')
ax.set_title('Estimation of optimal Beta L value')

ax.legend()
f.savefig('..//figures//dqn_ddqn_model1_estimation_Beta_L.png')
```
